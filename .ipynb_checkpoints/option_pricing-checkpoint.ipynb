```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import lightgbm as lgb
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RandomizedSearchCV,KFold

from tensorflow.keras import layers, models
import tensorflow as tf
```

```python
SEED = 42
np.random.seed(SEED)
```

```python
df = pd.read_excel("option_prices_data.xlsx").dropna()
df
```

```python
symbols = df["symbol"].unique()
TRAIN_SIZE = 0.8

train_df = pd.DataFrame()
test_df = pd.DataFrame()

for symbol in symbols:
    section = df[df["symbol"] == symbol]
    l = len(section)

    train_df = pd.concat([train_df,section.iloc[:int(l*TRAIN_SIZE),:]])
    test_df = pd.concat([test_df,section.iloc[int(l*TRAIN_SIZE):,:]])
```

```python
train_df
```

```python
test_df
```

```python
train_data = train_df[["strike_price", "equity_price", "annual_volatility", "riskfree_rate", "time_until_exp"]].values
tarin_labels = train_df["close_price"].values

test_data = test_df[["strike_price", "equity_price", "annual_volatility", "riskfree_rate", "time_until_exp"]].values
test_labels = test_df["close_price"].values
```

# Model training

## LightGBM

```python
data = train_df[["strike_price", "equity_price", "annual_volatility", "riskfree_rate", "time_until_exp"]].values
label = train_df["close_price"].values

# val = test_df[["strike_price", "equity_price", "annual_volatility", "riskfree_rate", "time_until_exp"]].values
# val_label = test_df["close_price"].values

train_data = lgb.Dataset(data, label=label)
# validation_data = lgb.Dataset(val, label=val_label)
```

```python
param = {
    # 'num_leaves': 18,
    'objective': 'mse',
    "learning_rate":0.02,
    "max_depth":25
    }
param['metric'] = 'mse'
```

```python
num_round = 1000
bst = lgb.train(param, train_data, num_round)
```

```python
bst
```

```python
bst.save_model('model.txt', num_iteration=bst.best_iteration)
```

```python
ypred = bst.predict(test_df[["strike_price", "equity_price", "annual_volatility", "riskfree_rate", "time_until_exp"]].values)
```

```python

print(mean_squared_error(ypred, test_df["close_price"].values))
print(mean_squared_error(test_df["BS"].values, test_df["close_price"].values))
```

## Neural Networks

```python
EPOCHS = 800
BATCH_SIZE = 256
VAL_SPLIT = 0.15
```

```python
scaler = StandardScaler()
train_data_sc = scaler.fit_transform(train_data)
test_data_sc = scaler.transform(test_data)
```

```python
model_1 = models.Sequential(
    [
        layers.Dense(256, activation="relu", input_shape=[train_data.shape[1]]),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(1, activation="relu"),

    ]
)

model_2 = models.Sequential(
    [
        layers.Dense(256, activation="relu", input_shape=[train_data.shape[1]]),
        layers.BatchNormalization(),
        layers.Dropout(0.15),
        layers.Dense(128, activation="relu"),
        layers.BatchNormalization(),
        layers.Dropout(0.15),
        layers.Dense(64, activation="relu"),
        layers.BatchNormalization(),
        layers.Dropout(0.15),
        layers.Dense(1, activation="relu"),
    ]
)

model_1.compile(
    optimizer = "adam",
    loss = "mse",
    metrics = [tf.keras.metrics.RootMeanSquaredError()]
  )

model_2.compile(
    optimizer = "adam",
    loss = "mse",
    metrics = [tf.keras.metrics.RootMeanSquaredError()]
  )
```

```python
# tarining model 1

history_1 = model_1.fit(
    x=train_data_sc,
    y=tarin_labels,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    verbose="auto",
    validation_split=VAL_SPLIT
)
```

```python
history_2 = model_2.fit(
    x=train_data_sc,
    y=tarin_labels,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    verbose="auto",
    validation_split=VAL_SPLIT
)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(list(range(EPOCHS)), history_1.history["loss"], label="train")
plt.plot(list(range(EPOCHS)), history_1.history["val_loss"], label="val")
plt.title("Model 1 - Loss (MSE)")
plt.legend()

plt.subplot(2,2,2)
plt.plot(list(range(EPOCHS)), history_1.history["root_mean_squared_error"], label="train")
plt.plot(list(range(EPOCHS)), history_1.history["val_root_mean_squared_error"], label="val")
plt.title("Model 1 - RMSE")
plt.legend()

plt.subplot(2,2,3)
plt.plot(list(range(EPOCHS)), history_2.history["loss"], label="train")
plt.plot(list(range(EPOCHS)), history_2.history["val_loss"], label="val")
plt.title("Model 2 - Loss (MSE)")
plt.legend()

plt.subplot(2,2,4)
plt.plot(list(range(EPOCHS)), history_2.history["root_mean_squared_error"], label="train")
plt.plot(list(range(EPOCHS)), history_2.history["val_root_mean_squared_error"], label="val")
plt.title("Model 2 - RMSE")
plt.legend()


plt.tight_layout()
```

```python
pred_1 = model_1.predict(test_data_sc)
pred_2 = model_2.predict(test_data_sc)
```

```python
print("model 1 (w/o dropout) RMSE:\t",mean_squared_error(pred_1, test_labels)**0.5)
print("model (with dropout)RMSE:\t",mean_squared_error(pred_2, test_labels)**0.5)
print("BS RMSE:\t\t",mean_squared_error(test_df["BS"], test_labels)**0.5)
```

```python
print("Neural Network MSE:",round(mean_squared_error(pred_2, test_labels),3))
print("Black Sholes MSE:",round(mean_squared_error(test_df["BS"], test_labels),3))
```
